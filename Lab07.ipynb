```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
df  = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
type(cancer)
```

```python
np.sum(pd.isnull(df).sum())
```

```python
cancer['target']
cancer['target'].shape
```

```python
df['cancer'] = pd.DataFrame(cancer['target'])
df.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='cancer',hue='mean radius',data=df,palette='rainbow')
```

```python
l=list(df.columns[0:10])
for i in range(len(l)-1):
  sns.boxplot(x='cancer',y=l[i],data=df,palette='winter')
  plt.title("Box plots of 10 cancer features ")
  plt.figure()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='cancer',hue='mean radius',data=df,palette='rainbow')
```

```python
f,(ax1,ax2) = plt.subplots(1,2,sharey=True , figsize=(12,6))
ax1.scatter(df['mean area'],df['cancer'])
ax1.set_title("cancer feature mean area ",fontsize = 15)
ax2.scatter(df['mean perimeter'],df['cancer'])
ax2.set_title("cancer feature mean perimeter ",fontsize = 10)
```
